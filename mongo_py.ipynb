### Loading collection (table) from mongodb to python

```python
import pymongo
import json
import pandas as pd
```

```python
client = pymongo.MongoClient("mongodb://localhost:27017")
db = client["acme"]
stock = db["post"]
```

```python
df =pd.DataFrame(list(stock.find()))
df
```

### To split the list and dict.

```python
df['user'].apply(pd.Series)
```

```python
df['tags'].apply(pd.Series)
```

```python
df1 = pd.concat([df.drop(['user'], axis=1), df['user'].apply(pd.Series)], axis=1)
```

```python
df2=pd.concat([df1.drop(['tags'], axis=1), df1['tags'].apply(pd.Series)], axis=1)
df2
```

### Rename column

```python
df3 = df2.rename(columns={0:"tag1"})
```

```python
df4 = df3.rename(columns={1:"tag2"})
df4
```

```python
#oops! we now have 2 columns with the same discription
# will drop all columns with nan, this will help eliminate it and since it is the only column with nan all through
df5 = df4.dropna(how='all', axis=1)
df5
```

```python
df5.drop(["_id","date"], axis=1, inplace=True)
df5
```
